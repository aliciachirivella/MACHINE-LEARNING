juntamos todos los excels en un solo dataframe y añadimos columna de año 

```python

# Ruta a la carpeta donde están los archivos
ruta_archivos = "excels/*.xlsx"  

dataframes = []
#glob.glob busca archivos con extensión .xlsx dentro de una carpeta llamada excels en el directorio actual.

for archivo in glob.glob(ruta_archivos):
    df = pd.read_excel(archivo)
    
    # Extraer el año del nombre del archivo
    año = archivo.split("_")[-1].replace(".xlsx", "")  # Supone nombres tipo "ventas_21.xlsx"
    df["año"] = int(año)  # Añadir el año como una columna
    dataframes.append(df)

df_total = pd.concat(dataframes, ignore_index=True)
```

He observado que los articulo se distinguen:
1. los que empiezan por 0 se diferencian si tienen los 2 ultimos numeros distintis
2. el resto se diferencian por los ultimos 3 

```python
#FUNCION PARA IDENTIFICAR CUAL ES 

def extraer_prefijo_sufijo(codigo):
    if codigo.startswith('00'):
        # Si comienza por "00", prefijo son los primeros 10 caracteres y sufijo los últimos 2
        prefijo = codigo[:10]
        sufijo = codigo[-2:]
    else:
        # Si no comienza por "00", prefijo son los primeros 7 caracteres y sufijo los últimos 3
        prefijo = codigo[:7]
        sufijo = codigo[-3:]
    return prefijo, sufijo

# Aplicar la función a la columna 'Artículo'
ventas[['Prefijo', 'Sufijo']] = ventas['Artículo'].apply(
    lambda x: pd.Series(extraer_prefijo_sufijo(x))
)

print(ventas[['Artículo', 'Prefijo', 'Sufijo']].head(2))
```

```python
# Función auxiliar para ejecutar consultas SQL
def sql_query(query):
    """
    Ejecuta una consulta SQL sobre la conexión SQLite y devuelve un DataFrame.
    """
    return pd.read_sql_query(query, connection)
```
